```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class Deque:
    def __init__(self):
        self.tail = None
        self.head = None
    
    def empty(self):
        if self.head == None:
            return True
        else:
            return False
    
    def addFirst(self, data):
        newNode = Node(data)
        newNode.next = self.head
        if self.empty():
            self.tail = newNode
        else:
            self.head.prev = newNode
        
        newNode.prev = None
        self.head = newNode
        
    def addLast(self, data):
        newNode = Node(data)
        newNode.prev = self.tail
        
        if self.empty():
            self.head = newNode
        else:
            self.tail.next = newNode
        
        newNode.next = None
        self.tail = newNode
        
    def removeFirst(self):
        removeNode = self.head
        if self.empty():
            return False
        removeData = self.head.data
        self.head = self.head.next
        removeNode = None
        
        if self.head == None:
            self.tail = None
        else:
            self.head.prev = None
    
        return removeData
        
    def removeLast(self):
        removeNode = self.tail
        if self.empty():
            return False
        removeData = self.tail.data
        self.tail = self.tail.prev
        removeNode = None

        if self.tail == None:
            self.head = None
        else:
            self.tail.next = None

        return removeData

    def getFirst(self):
        if self.empty():
            return False
        return self.head.data
    
    def getLast(self):
        if self.empty():
            return False
        return self.tail.data
```

```python
dq = Deque()
dq.addFirst(3)
dq.addFirst(2)
dq.addFirst(1)

dq.addLast(4)
dq.addLast(5)
dq.addLast(6)

while(dq.empty()==False):
    print(dq.removeFirst())

dq.addFirst(3)
dq.addFirst(2)
dq.addFirst(1)

dq.addLast(4)
dq.addLast(5)
dq.addLast(6)
while(dq.empty()==False):
    print(dq.removeLast())
```
